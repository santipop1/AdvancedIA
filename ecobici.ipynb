```python
import requests
import pandas as pd
import matplotlib as plt
import seaborn as sns


url ='https://ecobici.cdmx.gob.mx/wp-content/uploads/2025/08/2025-07.csv'
local_filename = '2025-07.csv'

try:
  response = requests.get(url, timeout=1200)
  response.raise_for_status() # Raise an exception for bad status codes

  with open(local_filename, 'wb') as f:
    f.write(response.content)

  print(f"Archivo '{local_filename}' descargado exitosamente.")

except requests.exceptions.RequestException as e:
  print(f"Error al descargar el archivo: {e}")
```

```python
dataframe = pd.read_csv(local_filename)
dataframe.count()
```

# Transformación

```python
df = dataframe.copy()

print(df[['Fecha_Retiro', 'Hora_Retiro', 'Fecha_Arribo', 'Hora_Arribo']].dtypes)
```

```python
df['Fecha_Retiro'] = pd.to_datetime(df['Fecha_Retiro'], dayfirst=True)
df['Fecha_Arribo'] = pd.to_datetime(df['Fecha_Arribo'], dayfirst=True)
```

```python
df['Hora_Arribo'] = pd.to_timedelta(df['Hora_Arribo'].astype(str))
df['Hora_Retiro'] = pd.to_timedelta(df['Hora_Retiro'].astype(str))
```

```python
df.info()
```

```python
# Verificar valores nulos en todas las columnas
print("\nConteo de valores nulos por columna:")
print(df.isnull().sum())
# genero usuario y edad con nullos


# Visualizar la distribución de la variable categórica 'genero'
print("\nVisualización de la distribución de 'genero':")

# Imprimir la cantidad de cada variable categórica en 'genero'
print("\nConteo de viajes por género:")
print(df['Genero_Usuario'].value_counts())
```

```python
# eliminar filas que tengan nulo el genero
df = df.dropna(subset=['Genero_Usuario'])

# llenar los valores nulos de edad con la mediana
df['Edad_Usuario'] = df['Edad_Usuario'].fillna(df['Edad_Usuario'].median())

```

```python
# Verificar valores nulos en todas las columnas
print("\nConteo de valores nulos por columna:")
print(df.isnull().sum())
```

```python
# juntar categoria 'O' con '?' en genero
df['Genero_Usuario'] = df['Genero_Usuario'].replace('O', '?')

print("\nConteo de viajes por género:")
print(df['Genero_Usuario'].value_counts())
```

```python
# agregamos columna duracion del viaje (fecha y hora de arribo - fecha y hora de retiro)
df["Arribo_dt"] = df["Fecha_Arribo"] + df["Hora_Arribo"]
df["Retiro_dt"] = df["Fecha_Retiro"] + df["Hora_Retiro"]

# Calcular la duración
df["Duracion"] = df["Arribo_dt"] - df["Retiro_dt"]
df["Duracion_minutos"] = df["Duracion"].dt.total_seconds() / 60


```
