```python
path = 'drive/MyDrive/TEC/7to semestre/Datasets/sf_housing_dirty.csv'
```

```python
import pandas as pd

df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
if df['price'].dtype == object:
  df['price'] = df['price'].str.replace('$', '')
  df['price'] = df['price'].str.replace(',', '')
  df['price'] = pd.to_numeric(df['price'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Matriz de correlación
corr_housing = df.select_dtypes(include=np.number).corr()

# Mostramos la matriz
plt.figure(figsize=(10,6))
sns.heatmap(corr_housing, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación - California Housing")
plt.show()
```
