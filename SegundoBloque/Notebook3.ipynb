```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
```

```python
# Generate a synthetic dataset
X, y = make_regression(n_samples=5000, n_features=1, noise=20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

y_pred = linear_model.predict(X_test)
```

```python
residuals = y_test - y_pred

plt.scatter(X_test, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('X_test')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
#
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_test.flatten(), y=y_test, label='Actual', color='blue')
sns.scatterplot(x=X_test.flatten(), y=y_pred, label='Predicted', color='red')
plt.xlabel('X_test')
plt.ylabel('y')
plt.title('Actual vs. Predicted Values')
plt.legend()
plt.show()
```

```python
# Plot: Homoscedasticity
plt.figure(figsize=(5, 5))
sns.scatterplot(x=y_pred, y=np.abs(residuals))
plt.xlabel('Predicted Values')
plt.ylabel('Absolute Residuals')
plt.title('Homoscedasticity Plot')
plt.show()
```

```python
# Generate dataset
np.random.seed(42)
X_example = np.random.rand(200, 1) * 100
y_example = 2 * X_example.squeeze() + np.random.normal(0,X_example.squeeze(), 200)

# Split in test and train
X_train, X_test, y_train, y_test = train_test_split(X_example, y_example, test_size=0.3, random_state=42)

# Train a model
model_h = LinearRegression()
model_h.fit(X_train, y_train)
y_pred_h = model_h.predict(X_test)
```

```python
residuals_h = y_test - y_pred_h
plt.figure(figsize=(5, 5))
sns.scatterplot(x=y_pred_h, y=np.abs(residuals_h))
plt.xlabel('Predicted Values')
plt.ylabel('Absolute Residuals')
plt.title('Homoscedasticity Plot')
plt.show()
```

#Dataset Wine

```python
# Cargar el dataset desde la URL oficial de UCI.
def cargar_dataset_vinos(url: str) -> pd.DataFrame:

    # pd.read_csv con delimiter=';' porque el archivo usa punto y coma como separador.
    return pd.read_csv(url, delimiter=';')

# URL oficial del dataset (vino tinto)
fuente_uci = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

# Cargamos el dataset en un DataFrame llamado 'vinos_df'
vinos_df = cargar_dataset_vinos(fuente_uci)

# Vista rpida de las primeras filas


vinosX = vinos_df.drop('quality', axis=1)
vinosY = vinos_df['quality']

vinosY.head()
```

```python
XTrain, XTest, yTrain, yTest = train_test_split(vinosX,vinosY, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(XTrain, yTrain)

y_pred = linear_model.predict(XTest)
```

```python
residuals = yTest - y_pred

plt.scatter(XTest['fixed acidity'], residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Fixed Acidity from XTest')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
