```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("blastchar/telco-customer-churn")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# The path variable from the previous cell contains the directory where the dataset files are located
# List the files in the directory to find the CSV file
dataset_files = os.listdir(path)
print("Files in the dataset directory:", dataset_files)

# Assuming the CSV file is named 'WA_Fn-UseC_-Telco-Customer-Churn.csv' based on the dataset name
csv_file_path = os.path.join(path, 'WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
display(df.head())
```

```python
display(df.head())
```

```python
df.info()
```

```python
#totalcharge to float
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Chrun to boolean
df['Churn'] = df['Churn'].map({'Yes': True, 'No': False})
```

```python
df.describe()
```

```python

#median of ternure
print(df['tenure'].median())

#median of MonhlyCharges
print(df['MonthlyCharges'].median())

#median of TotalCharges
print(df['TotalCharges'].median())
```

# 2. Distribution Visualization

```python
# Distribution
import matplotlib.pyplot as plt
import seaborn as sns

# histogram of ternure
plt.figure(figsize=(8, 6))
sns.histplot(df['tenure'], bins=20, kde=True)
plt.title('Distribution of Tenure')
plt.xlabel('Tenure in months')
plt.ylabel('Frequency')
plt.show()
```

```python
#histogram of Monthly Charges
plt.figure(figsize=(8, 6))
sns.histplot(df['MonthlyCharges'], bins=20, kde=True)
plt.title('Distribution of Monthly Charges')
plt.xlabel('Monthly Charges')
plt.ylabel('Frequency')
plt.show()
```

```python
#histogram of Total Charges
plt.figure(figsize=(8, 6))
sns.histplot(df['TotalCharges'], bins=20, kde=True)
plt.title('Distribution of Total Charges')
plt.xlabel('Total Charges ')
plt.ylabel('Frequency')
plt.show()
```
